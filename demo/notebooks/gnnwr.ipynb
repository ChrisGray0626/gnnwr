```python
from gnnwr import models,datasets
import pandas as pd
```

```python
data = pd.read_csv(u'../../data/simulated_data.csv')
```

```python
train_dataset, val_dataset, test_dataset = datasets.init_dataset(data=data,
                                                        test_ratio=0.2,
                                                        valid_ratio=0.1,
                                                        x_column=['x1', 'x2'],
                                                        y_column=['y'],
                                                        spatial_column=['u', 'v'],
                                                        sample_seed=42)
```

```python
SGD_PARAMS = {
    "maxlr": 1,
    "minlr": 0.1,
    "upepoch": 10000,
    "decayepoch": 60000,
    "decayrate": 0.98,
}
gnnwr = models.GNNWR(train_dataset, val_dataset, test_dataset, [32, 16, 8])
```

```python
gnnwr.run(100)
```
